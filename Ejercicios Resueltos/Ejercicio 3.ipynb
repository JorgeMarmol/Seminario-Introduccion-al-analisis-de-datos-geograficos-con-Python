```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://ourairports.com/data/airports.csv"
df = pd.read_csv(url)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="type", y="elevation_ft")
plt.title("Elevacin por tipo de aeropuerto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="continent", order=df["continent"].value_counts().index)
plt.title("Nmero de aeropuertos por continente")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["elevation_ft"].dropna(), bins=50, kde=True)
plt.title("Distribucin de elevaciones de aeropuertos")
plt.xlabel("Elevacin (pies)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="longitude_deg", y="latitude_deg", hue="type", alpha=0.5, s=20)
plt.title("Ubicacin de aeropuertos por tipo")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.tight_layout()
plt.show()
```

```python
top10_paises = df["iso_country"].value_counts().head(10)

plt.figure(figsize=(8, 5))
sns.barplot(x=top10_paises.index, y=top10_paises.values)
plt.title("Top 10 pases con ms aeropuertos")
plt.xlabel("Pas")
plt.ylabel("Cantidad de aeropuertos")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x="continent", y="elevation_ft", inner="quartile", scale="width", palette="muted")
plt.title("Distribucin de elevacin por continente (Violin Plot)")
plt.xlabel("Continente")
plt.ylabel("Elevacin (pies)")
plt.ylim(-1000, 10000)  # recorta para evitar outliers extremos visualmente molestos
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=df, x="type", y="elevation_ft", jitter=0.3, alpha=0.3)
plt.title("Elevacin por tipo de aeropuerto (Strip Plot con Jitter)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
